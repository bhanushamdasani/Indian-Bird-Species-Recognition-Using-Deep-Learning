```python
import os
import librosa
import noisereduce as nr
import soundfile as sf

input_folder = 'C:\\Users\\Admin\\Desktop\\audios\\Rashika_-_Rufous_Treepie'
output_folder = 'cleaned_Rufous_Treepie'
os.makedirs(output_folder, exist_ok=True)

# Get a list of all files in the input folder
audio_files = [f for f in os.listdir(input_folder) if f.endswith('.wav')]

# Iterate over each audio file in the folder
for audio_file in audio_files:
    # Construct the full path for input and output files
    input_audio_file = os.path.join(input_folder, audio_file)
    output_audio_file = os.path.join(output_folder, f"cleaned_{audio_file}")

    try:
        # Load the input audio file
        audio_signal, sample_rate = librosa.load(input_audio_file, sr=None)

        # Apply noise reduction
        reduced_audio = nr.reduce_noise(y=audio_signal, sr=sample_rate)

        # Save the cleaned audio as a WAV file
        sf.write(output_audio_file, reduced_audio, sample_rate)

        print(f"Cleaned audio saved to {output_audio_file}")
    except Exception as e:
        print(f"Error processing {audio_file}: {e}")
```
